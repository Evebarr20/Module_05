# Regex Questions

For each example below find a single regex expression that will extract or find all instances of the pattern

Show that it works in each case,  with no false detections

```python
import re
```

## Problem 1
Using the text below,  extract all e-mail addresses from the three lines of text


```python
text = "Contact us at support@example.com or sales@company.net."
text2 ="Heat @ 200 C for 3 minutes"
text3="We we purchaes 200 units@ $40 each,  let me know at if that works, I'm SmithW@dubious.com"

regex = "[A-Za-z0-9]+@[A-Za-z]+\.[A-Za-z]{3}"
re.findall(regex, text)
```

```python
re.findall(regex, text2)
```

```python
re.findall(regex, text3)
```

## Problem 2

Extract years in Month/Day/Year format

dashes or slashes should be okay

```python
text="The event will be held on 10-27-2023."
text2="Today is 7/24/2025"
text3="The part number is 123-1-3785"

date_regex = r"\b\d{1,2}[-/]\d{1,2}[-/]\d{4}\b"
re.findall(date_regex, text)
```

```python
re.findall(date_regex, text2)
```

```python
re.findall(date_regex, text3)
```

## Problem 3

Find the string that starts with an upper or lower case A (or a)

```python
text="Bob is out sick today"
text2="again, Bob is out sick"
text3="Aargh, we will be short-handed for the dinner rush"
case_regex = "^[Aa].*"
re.findall(case_regex, text)
```

```python
re.findall(case_regex, text2)
```

```python
re.findall(case_regex, text3)
```

## Problem 4

Write a single regex operation that removes all the excess spaces from the text below

```python
text="    Hello    Delaware  !!"
text2="Were is  Bob?"
text3=" How  did we get      here"

remove_regex = '\s{2,}'
re.sub(remove_regex,' ', text).strip()
```

```python
re.sub(remove_regex,' ', text2).strip()
```

```python
re.sub(remove_regex,' ', text3).strip()
```

# Problem 5

Extract the full web address

```python
text="check out the website at https://WiggleWare.com"
text2="The reference site is http://lotsa_bugs.edu"

web_regex = "https?://[a-zA-z_]+\.[A-Za-z]{3}"
re.findall(web_regex, text)
```

```python
re.findall(web_regex, text2)
```
